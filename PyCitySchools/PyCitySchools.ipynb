```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
TotalSchools = len(school_data_complete["School ID"].unique())
TotalSchools
```

```python
TotalStudents = len(school_data_complete["Student ID"])
TotalStudents
```

```python
TotalBudget = sum(school_data["budget"])
TotalBudget
```

```python
AverageMath = school_data_complete["math_score"].mean()
AverageMath
```

```python
AverageReading = school_data_complete["reading_score"].mean()
AverageReading
```

```python
#percentage of math scores that passed
passingmath = school_data_complete.loc[school_data_complete["math_score"] >=70]["math_score"].count()/TotalStudents
passingmath = passingmath * 100
```

```python
#percentage of reading scores that passed
passingreading = school_data_complete.loc[school_data_complete["reading_score"] >=70]["reading_score"].count()/TotalStudents
passingreading = passingreading * 100
```

```python
St_passing = school_data_complete.loc[(school_data_complete["reading_score"] >=70) & (school_data_complete["math_score"] >=70)]['student_name'].count()/TotalStudents
St_passing = St_passing * 100
```

```python
districtsummary_table = pd.DataFrame({"Total Schools": [TotalSchools], "Total Students": [TotalStudents], "Total Budget": [TotalBudget], "Average Math Score": [AverageMath], "Average Reading Score": [AverageReading], "% Passing Math": [passingmath], "% Passing Reading": [passingreading], "% Overall Passing": [St_passing]})
districtsummary_table
```

```python
districtsummary_table ['Total Budget'] = districtsummary_table ['Total Budget'].map("${:}".format)
districtsummary_table ['Average Math Score'] = districtsummary_table ['Average Math Score'].map("{:.2f}%".format)
districtsummary_table ['Average Reading Score'] = districtsummary_table ['Average Reading Score'].map("{:.2f}%".format)
districtsummary_table ['% Passing Math'] = districtsummary_table ['% Passing Math'].map("{:.2f}%".format)
districtsummary_table ['% Passing Reading'] = districtsummary_table ['% Passing Reading'].map("{:.2f}%".format)
districtsummary_table ['% Overall Passing'] = districtsummary_table ['% Overall Passing'].map("{:.2f}%".format)
districtsummary_table
#end of District Summary
```

```python
#Start of School Summary
School_Cat = school_data_complete.set_index('school_name')
School_Cat = School_Cat.groupby(['school_name'])
School_Cat.head()
```

```python
School_Types = school_data.set_index('school_name')['type']
School_Types
```

```python
T_Students = School_Cat["Student ID"].count()
T_Students
```

```python
School_Budget = school_data.set_index('school_name')['budget']
School_Budget
```

```python
School_Budget_St = School_Budget / T_Students
School_Budget_St
```

```python
AvgReading = School_Cat ['reading_score'].mean()
AvgReading
```

```python
AvgMath = School_Cat ['math_score'].mean()
AvgMath
```

```python
PassedReading = school_data_complete.loc[school_data_complete["reading_score"] >=70].groupby("school_name")["student_name"].count()/T_Students
PassedReading = PassedReading *100
PassedReading
```

```python
PassedMath = school_data_complete.loc[school_data_complete["math_score"] >=70].groupby("school_name")["student_name"].count()/T_Students
PassedMath = PassedMath*100
PassedMath
```

```python
PassedBoth = school_data_complete.loc[(school_data_complete["reading_score"] >=70) & (school_data_complete["math_score"] >=70)].groupby("school_name")["student_name"].count()/T_Students
PassedBoth = PassedBoth *100
PassedBoth
```

```python
#Putting a Dataframe together....no []
schoolsummary_table = pd.DataFrame({"School Type": School_Types,"Total Students": T_Students, "Total School Budget": School_Budget, "Per Student Budget": School_Budget_St, "Average Math Score": AvgMath, "Average Reading Score": AvgReading, "% Passing Math": PassedMath, "% Passing Reading": PassedReading, "% Overall Passing": PassedBoth})
schoolsummary_table
```

```python
#Giving int() symbols for readability
schoolsummary_table ['Total School Budget'] = schoolsummary_table ['Total School Budget'].map("${:}".format)
schoolsummary_table ['Per Student Budget'] = schoolsummary_table ['Per Student Budget'].map("${:}".format)
schoolsummary_table ['Average Math Score'] = schoolsummary_table ['Average Math Score'].map("{:.2f}%".format)
schoolsummary_table ['Average Reading Score'] = schoolsummary_table ['Average Reading Score'].map("{:.2f}%".format)
schoolsummary_table ['% Passing Math'] = schoolsummary_table ['% Passing Math'].map("{:.2f}%".format)
schoolsummary_table ['% Passing Reading'] = schoolsummary_table ['% Passing Reading'].map("{:.2f}%".format)
schoolsummary_table ['% Overall Passing'] = schoolsummary_table ['% Overall Passing'].map("{:.2f}%".format)
schoolsummary_table
```

```python
#Bottom 5 Schools Based on % Overall Passing    #.sort_values
Bottom5 = schoolsummary_table.sort_values("% Overall Passing",ascending =True)
Bottom5
```
